```python
import json
import numpy as np
import os
import SimpleITK as sitk
from collections import defaultdict
```

```python
with open('/home/t722s/Desktop/Sam-Med3DTest/BratsMiniOld/multi_run_results_original.json', 'r') as f:
    resd = json.load(f)

image_IMAGE = sitk.ReadImage('/home/t722s/Desktop/Datasets/BratsMini/imagesTs/BraTS2021_01646.nii.gz')
mask_IMAGE = sitk.ReadImage('/home/t722s/Desktop/Datasets/BratsMini/labelsTs/BraTS2021_01646.nii.gz')
im = sitk.GetArrayFromImage(image_IMAGE)
gt = sitk.GetArrayFromImage(mask_IMAGE)
clicks_p = defaultdict(list)
clicks_n = defaultdict(list)

for k,v in resd.items():
    for i in range(5):
        pt, lb = v['points'][i], v['labels'][i]
        clicks_p[i+1].append(pt) if lb else clicks_n[i+1].append(pt)

for i in range(5):
    clicks_p[i+1] = np.array(clicks_p[i+1])
    clicks_n[i+1] = np.array(clicks_n[i+1])

points_mask_p = dict()
points_mask_n = dict()
for i in range(5):
    points_mask_p[i+1] = np.zeros_like(gt)
    points_mask_p[i+1][*clicks_p[1].T] = 255
    points_mask_n[i+1] = np.zeros_like(gt)
    points_mask_n[i+1][*clicks_n[1].T] = 255
```

```python
dest = '/home/t722s/Desktop/Sam-Med3DTest/BratsMini'
for i in range(5):
    f = os.path.join(dest, f'clicks_p_{i+1}.nii.gz')
    mask: sitk.Image = sitk.GetImageFromArray(points_mask_p[i+1])
    mask.CopyInformation(image_IMAGE)
    sitk.WriteImage(mask, f)

    f = os.path.join(dest, f'clicks_n_{i+1}.nii.gz')
    mask = sitk.GetImageFromArray(points_mask_n[i+1])
    mask.CopyInformation(image_IMAGE)
    sitk.WriteImage(mask, f)
```
